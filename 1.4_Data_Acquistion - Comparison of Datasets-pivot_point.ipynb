# Comparing my Data to Spotify's Skip Challenge
This is my moment of truth notebook. I have now gathered all the information I need to compare my data to the Skip Challenge's Dataset.
I will compare rows and find how many match. If I have any matches here I will then gather the track_id from the Skip Challenge's Track Features Dataset and match it to Skip Challenge's Testing/Training dataset, where the track ID may appear multiple times.

### Imports

```python
import pandas as pd
import numpy as np
```

```python
tracks = pd.read_csv('./data/tracks_and_features.csv') # my gathered data
tf1 = pd.read_csv('./data/tf_000000000000.csv') # CrowdAI's Skipp Challenge Track Features
tf2 = pd.read_csv('./data/tf_000000000001.csv')
```

### Moment of Truth
Before I go through and change column values to be in the same format accross the board, I will change the duration column (my dataframe is in ms. and skip challenge is in s.) and see how many match. From my gathered dataset I need to drop all tracks that were released later than 2018, becuase the Skip Challenge goes up to 2018.

```python
tracks.drop(tracks.loc[tracks['release_year'] > 2018].index, inplace=True)
```

```python
tf1.head()
```

```python
# Changing duration ms to s and then dropping the duration_ms column
tracks['duration'] = tracks['duration_ms'] / 1000
tracks.drop('duration_ms', axis=1, inplace = True)
```

```python
print(f" There are {tracks['duration'].isin(tf1['duration']).value_counts()[1] + tracks['duration'].isin(tf2['duration']).value_counts()[1]} rows total that match the duration of my dataset") 
```

## Conclusion
That is a big drop of information. The matching numbers will continue to go down as I match other columns, which is necessary because I cannot confidently say that the track that was skipped was in-fact the one I identified based on duration alone.
